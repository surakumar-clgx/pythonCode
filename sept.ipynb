```python
import os
import pandas as pd
from concurrent.futures import ThreadPoolExecutor

from time import perf_counter,sleep
start = perf_counter()

# This is the input  folder
path=r"\\filer-argentina-dev\Argentina\Shared Folder\Manikanda Gokul\Sept\\"

file_name= [file for file in os.listdir(path) if file.startswith('sept')]

#for i in file_name
#    print("vis","s.no ",i,"  ",file_name[i])

pd1 = pd.DataFrame()
p =1
for i in file_name:
    print(i)
    #vishnu excelfile
    #print("v - ",i)
    xls = pd.ExcelFile( rf"{path}{i}")

    n = pd.read_excel(xls,  sheet_name = "field_level_data")
    pd1 = pd.concat([pd1,n], axis=0)
    #print("v -",pd1)
    print(f"{p}---",n.shape)
    print(pd1.shape)
    p+=1

print(pd1.shape)

end = perf_counter()
        
print("time---",end-start)
```
