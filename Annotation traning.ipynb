```python
data = []
files_annotated = []
for annotFile in Path(annot_path).iterdir():
    print(annotFile)
    annots = None
    with open(annotFile, "r") as file:
        annots = json.load(file)
    print(len(annots))
    for annot in annots:
        ocrName = annot["file_upload"].split("-")[1]
        #         print(ocrName)
        if ocrName in files_annotated:
            continue
        files_annotated.append(ocrName)
        text = None
        with open(f"{ocrs_path}/{ocrName}", "r", encoding="utf8") as f:
            text = f.read()
        results = annot["annotations"][0]["result"]
        #         print(len(results))
        ent_dict = {"entities": []}
        for entities in results:
            ent = entities["value"]
            ent["labels"] = ",".join(ent["labels"])
            print(tuple(ent.values()))
            ent_dict["entities"].append(tuple(ent.values()))
        ocr_ent_combined = (text, ent_dict)
        data.append(ocr_ent_combined)
#         break
#     break
```

```python
import random

import spacy
from spacy.training import Example
```

```python
nlp = spacy.blank("en")  # Create a new, blank English model
if "ner" not in nlp.pipe_names:
    ner = nlp.add_pipe("ner", last=True)
else:
    ner = nlp.get_pipe("ner")
```

```python
for text, annotations in data:
    for ent in annotations.get("entities"):
        ner.add_label(ent[2])
```

```python
TRAIN_DATA, TEST_DATA = data[:15], data[15:]
```

```python
# Disable other pipeline components during training
other_pipes = [pipe for pipe in nlp.pipe_names if pipe != "ner"]

with nlp.disable_pipes(*other_pipes):
    optimizer = nlp.begin_training()
    for itn in range(10):  # Number of training iterations
        random.shuffle(TRAIN_DATA)
        losses = {}
        for text, annotations in TRAIN_DATA:
            doc = nlp.make_doc(text.upper())
            spans = []
            ent = annotations["entities"]
            for e in ent:
                start, end, label = e[0], e[1], e[2]
                span = doc.char_span(start, end, label, alignment_mode="expand")
                spans.append(span)
            spans = filter_spans(spans)
            example = Example.from_dict(doc, {"entities": spans})
            nlp.update([example], drop=0.5, losses=losses)
        print(losses)
```
